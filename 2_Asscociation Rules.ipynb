## **Phần giải toán** 
### **Câu (a):**
Xem xét luật kết hợp A → B, ta có: 

>>$Conf(A → B) = \frac{S(A ∩ B)}{S(A)}$

Vì độ tin cậy (`Confidence`) bỏ qua $S(B)$ - tần suất xuất hiện của B trong tập dữ liệu. Đây là một sự hạn chế vì độ tin cậy cao có thể là do item B có tần suất xuất hiện lớn hơn item A, điều này sẽ khiến cho luật kết hợp trở nên thiên vị và sẽ không có ý nghĩa. 

***Ví dụ:*** Một siêu thị bán nhiều sữa hơn bánh mì, cụ thể là 100 đơn vị sữa và 20 đơn vị bánh mì thì ta luật kết hợp $bread → milk$ có công thức như sau:

>>$Conf(bread → milk) = \frac{S(bread ∩ milk)}{S(bread)} = \frac{20}{20} = 1$

Ví dụ trên cho thấy độ tin cậy cao là vì sữa thường xuất hiện trong tập dữ liệu, không phải là vì có mối liên hệ chặt chẽ giữa sữa và bánh mì.

Ngược lại, `Lift` và `Conviction` không có sự hạn chế này do có tính đến $S(B)$ trong công thức. `Lift` và `Conviction` so sánh xác suất xuất hiện đồng thời của 2 itemset trong luật kết hợp so với lại xác suất xuất hiện cùng nhau dưới điều kiện 2 itemset trong luật kết hợp độc lập với nhau. Do đó, `Lift` và `Conviction` không bị ảnh hưởng bởi tần suất xuất hiện của B trong tập dữ liệu. 

### **Câu (b):**

Trong các hệ đo lường được nêu trên, chỉ có `Lift` đối xứng.

Với $Lift(A → B)$, ta có :

>>$Lift(A \rightarrow B) = \frac{Conf(A \rightarrow B)}{S(B)} = \frac{S(A \cap B)}{S(A) \cdot S(B)} = \frac{S(B \cap A)}{S(B) \cdot S(A)} = Lift(B \rightarrow A)$

*→ Vì vậy, hệ số đo lường `Lift` đối xứng.*

Ví dụ về hệ đo lường không đối xứng. Ta có tập dữ liệu sau:

    {A, B}
    {A, C}
    {B, C}
    {A, B, C}
    {A, B, D}
    {B, C, D}
    {A, B, C, D}

Ta có: $S(A)=5,  S(B)=6, S(A ∩ B)=3$ 

- Đối với `Confidence`, ta có :

>>$Conf(A → B) = \frac{S(A ∩ B)}{S(A)} = \frac{3}{5} $

>>$Conf(B → A) = \frac{S(A ∩ B)}{S(B)} = \frac{3}{6} = \frac{1}{2} $

*→ Độ tin cậy không đối xứng trong trường hợp này. Do $Conf(A → B) ≠ Conf(B → A)$.*

- Đối với `Conviction`, ta có :

>>$Conv(A → B) = \frac{1 - S(B)}{1 - \frac{S(A ∩ B)}{S(A)}} = \frac{1 - 6/7}{1 - 3/5} = \frac{1}{2} $

>>$Conv(B → A) = \frac{1 - S(A)}{1 - \frac{S(A ∩ B)}{S(B)}} = \frac{1 - 5/7}{1 - 3/6} = \frac{3}{10} $

*→ `Conviction` không đối xứng trong trường hợp này. Do $Conv(A → B) ≠ Conv(B → A)$.*

### **Câu (c):**

Xét một luật kết hợp hoàn hảo A → B, nghĩa là mọi giao dịch chứa A đều chứa B. Trong trường hợp này, ta có $S(A ∩ B) = S(A) = S(B)$. Sử dụng các giá trị này, ta có:

- `Confidence`: 

>>$Conf(A → B) = \frac{S(A ∩ B)}{S(A)} = 1 $

- `Lift`: 

>>$Lift(A → B) = \frac{Conf(A → B)}{S(B)} = \frac{S(A ∩ B)}{S(A) \cdot S(B)} = \frac{1}{1} = 1$

- `Conviction`: 

>>$Conv(A → B) = \frac{1 - S(B)}{1-Conf(A → B)} = \frac{1 - S(B)}{1 - \frac{S(A ∩ B)}{S(A)}} = \frac{1 - 1}{1 - 1} = \frac{0}{0} = ∞$

Kết quả của hệ số `Conviction` là vô cực, vì 0/0 là không xác định. Tuy nhiên, nếu ta lấy giới hạn khi $Conf(A → B)$ tiến tới 1, thì $Conv(A → B)$ tiến tới vô cùng. Đây là giá trị tối đa có thể đạt được đối với `Conviction`. Do đó, `Conviction` có tính chất này.




## **Phần áp dụng**

**Đọc dữ liệu đầu vào**

```python
# Set minimum support
min_support = 100
```

```python
# Read data from file
input_path = 'browsing.txt'
dataset = []
with open(input_path, 'r') as f:
    # With each line in file
    for line in f:
        # Append a sorted set of items in line to dataset
        dataset.append(sorted(set(line.strip().split(' '))))
```

```python
# Print number of transactions
print('Number of transactions: {}'.format(len(dataset)))
```

**Chạy thuật toán Apriori để tìm các itemset thỏa mãn mức support tối thiểu trong lần pass đầu tiên**

```python
# Create a dictionary to store support of each item
singles_support = {}

# Loop through each transaction in dataset
for transaction in dataset:
    # For each item in transaction
    for item in transaction:
        # Increment support of item by 1
        singles_support[item] = singles_support.get(item, 0) + 1

# Filter out items that do not meet minimum support
singles_support = {k: v for k, v in singles_support.items() if v >= min_support}

# Print number of frequent items that meet minimum support threshold in the first pass
print('Number of frequent items in first pass: {}'.format(len(singles_support)))
```

### **Câu (d):**
**Chạy thuật toán Apriori để tìm các cặp itemset có độ dài bằng 2 trong lần pass thứ 2**

```python
# Create a dictionary to store support of each pair of items
double_support = {}

# With each transaction in dataset
for transaction in dataset:
    # For each item in transaction
    for i in range(len(transaction)):
        # Check if item meets minimum support threshold in the first pass
        if transaction[i] in singles_support:
            # If yes, check if any item before it meets minimum support threshold
            for j in range(i):
                # Check if item meets minimum support threshold in the first pass
                if transaction[j] in singles_support:
                    # If yes, we have found a pair of items that all meet minimum support threshold in the first pass
                    # Increment support of this pair by 1
                    double_support[(transaction[j], transaction[i])] = double_support.get((transaction[j], transaction[i]), 0) + 1

# Filter out pairs of items that do not meet minimum support
double_support = {k: v for k, v in double_support.items() if v >= min_support}

# Print number of frequent double items that meet minimum support threshold in the second pass
print('Number of frequent double items in second pass: {}'.format(len(double_support)))
```

**Xác định 10 luật kết hợp có confidence cao nhất**

```python
# Identify top 10 association rules with highest confidence
# All association rules are of the form X -> Y or Y -> X
doubles_conf = []
for double, support in double_support.items():
    x, y = double
    xy_conf = support / singles_support[x]
    yx_conf = support / singles_support[y]
    doubles_conf.extend([(((x), y), xy_conf), 
                         (((y), x), yx_conf)])

# Sort by confidence in descending order
# If two rules have the same confidence, sort by the item in the left side in association rule
doubles_conf = sorted(doubles_conf, key=lambda x: (-x[1], x[0][0][0]))

# Print top 10 association rules
print('Top 10 association rules:')
for i in range(10):
    print('conf({} -> {}) = {:.3f}'.format(doubles_conf[i][0][0], doubles_conf[i][0][1], doubles_conf[i][1]))
```

### **Câu (e):**
**Chạy thuật toán Apriori để tìm các cặp itemset có độ dài bằng 3 trong lần pass thứ 3**

```python
# Create a dictionary to store support of each triple of items
triple_support = {}

for transaction in dataset:
    for i in range(len(transaction)):
        if transaction[i] not in singles_support:
            continue
        for j in range(i):
            if transaction[j] not in singles_support:
                continue
            if (transaction[j], transaction[i]) not in double_support:
                continue
            for k in range(j):
                if transaction[k] not in singles_support:
                    continue
                if (transaction[k], transaction[j]) not in double_support:
                    continue
                if (transaction[k], transaction[i]) not in double_support:
                    continue
                # If all items in triple meet minimum support threshold in the first pass and in the second pass
                # Increment support of this triple by 1
                triple_support[(transaction[k], transaction[j], transaction[i])] = triple_support.get((transaction[k], transaction[j], transaction[i]), 0) + 1

# Filter out triples of items that do not meet minimum support
triple_support = {k: v for k, v in triple_support.items() if v >= min_support}

# Print number of frequent triple items that meet minimum support threshold in the third pass
print('Number of frequent triple items in third pass: {}'.format(len(triple_support)))
```

**Xác định 10 luật kết hợp có confidence cao nhất**

```python
# Identify top 10 association rules with highest confidence
# All association rules are of the form {X, Y} -> Z or {X, Z} -> Y or {Y, Z} -> X
triples_conf = []
for triple, support in triple_support.items():
    x, y, z = triple
    xy_z_conf = support / double_support[(x, y)]
    xz_y_conf = support / double_support[(x, z)]
    yz_x_conf = support / double_support[(y, z)]
    triples_conf.extend([(((x, y), z), xy_z_conf), 
                         (((x, z), y), xz_y_conf), 
                         (((y, z), x), yz_x_conf)])

# Sort by confidence in descending order
# If two rules have the same confidence, sort by the each item in the left side in association rule
triples_conf = sorted(triples_conf, key=lambda x: (-x[1], x[0][0][0], x[0][0][1]))

# Print top 10 association rules
print('Top 10 association rules:')
for i in range(10):
    print(f"conf({{{triples_conf[i][0][0][0]}, {triples_conf[i][0][0][1]}}} -> {triples_conf[i][0][1]}) = {triples_conf[i][1]:.3f}")
```
